```python
#creating an array to using it in getting bar plot for specific country Names
Required_Countries=["United States","United Kingdom","Europian Union","Germany","Bangladesh"]
 
#filtering DataFrame to Get required country Names
filtered_data=df[df["Country Name"].isin(Required_Countries)]
filtered_data
```

```python
#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```
